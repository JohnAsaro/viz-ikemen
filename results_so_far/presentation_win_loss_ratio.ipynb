```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv('presentation_model_test.csv')

# Count wins
win_counts = df['winner'].value_counts().sort_index()

# Extract counts
player1_wins = win_counts.get(1, 0)
player2_wins = win_counts.get(2, 0)

# Calculate percentages
total_games = player1_wins + player2_wins
player1_pct = (player1_wins / total_games) * 100 if total_games > 0 else 0
player2_pct = (player2_wins / total_games) * 100 if total_games > 0 else 0

# Print stats
print(f"Player 1 Wins: {player1_wins} ({player1_pct:.2f}%)")
print(f"Player 2 Wins: {player2_wins} ({player2_pct:.2f}%)")

# Plot bar chart
plt.bar(['AI', 'Sparring Partner'], [player1_wins, player2_wins], color=['blue', 'green'])
plt.ylabel('Number of Wins', fontsize=15, font='Times New Roman')
plt.title('Winrates', font='Times New Roman', fontsize=18)
plt.ylim(0, 700)
plt.text(0, player1_wins + 4, f'{player1_pct:.1f}%', ha='center', fontsize=13, font='Times New Roman')
plt.text(1, player2_wins + 4, f'{player2_pct:.1f}%', ha='center', fontsize=13, font='Times New Roman')
plt.tight_layout()
plt.show()
```
